```python
fig = plt.figure(figsize=(20, 6))
# 1. Heatmap
plt.subplot(1, 3, 1)
heatmap_data = pd.crosstab(df['health'], df['reduced'])
sns.heatmap(heatmap_data, annot=True, cmap='YlOrRd', fmt='d')
plt.title('Health Status vs Activity Reduction\nHeatmap')
plt.xlabel('Activity Reduction Level')
plt.ylabel('Health Status')
plt.show()
```

```python
fig = plt.figure(figsize=(20, 7))
# 2. Box Plot
plt.subplot(1, 3, 2)
sns.boxplot(x='health', y='reduced', data=df)
plt.title('Activity Reduction by Health Status\nBox Plot')
plt.xlabel('Health Status')
plt.ylabel('Activity Reduction Level')
plt.show
```

```python
fig = plt.figure(figsize=(20, 6))
# 3. Violin Plot
plt.subplot(1, 3, 3)
sns.violinplot(x='health', y='reduced', data=df)
plt.title('Activity Reduction by Health Status\nViolin Plot')
plt.xlabel('Health Status')
plt.ylabel('Activity Reduction Level')
```
